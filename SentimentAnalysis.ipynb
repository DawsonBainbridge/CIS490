##SentimentAnalysis

**sentiment analysis** - identify/extract *affective states*. At the minimal level involves detecting the positive or negative tone from text. 
Next level may be common emotions. 

[**VADER Valence Aware Dictionary for sEntiment Reasoning **]()

```python
#VADER
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

```

```python
sent1 = 'I LOVE BITCONNECT!'
sent2 = 'I love bitconnect.'
sent2_5 = 'I love bitconnect!'
sent3 = 'I DO NOT LIKE BITCONNECT'
sent4 = 'I hate your face, but i still love you'
print(sid.polarity_scores(sent1))
print(sid.polarity_scores(sent2))
print(sid.polarity_scores(sent2_5))
print(sid.polarity_scores(sent3))
print(sid.polarity_scores(sent4))
```

```python
nltk.download('book')
from nltk.book import *
```

```python
text1_str = ' '.join(list(text1))
text1_sent = nltk.sent_tokenize(text1_str)
```

```python
print(text1_sent[100:-100])
```

```python
pos = []
for s in text1_sent[100:1000]:
  pos.append(sid.polarity_scores(s)['pos'])
print(pos)
```

```python
import matplotlib.pyplot as plt
plt.plot(pos)
plt.show()
aves=[]
for i in range(int(len(pos)/10)):
  ave = sum(pos[i*10:10*i+10])/10
  aves.append(ave)

plt.plot(aves)
plt.show()
```
